```python
#3
#N_total
# Calculate Mean
mean = df['N_total'].mean()
# Calculate standard deviation
std = df['N_total'].std()
# Calculate quantile
quantile =df['N_total'].quantile()





print(f" Mean of N_total : {mean}")
print(f" standard deviation of N_total : {std}")
print(f" Quantile of N_total : {quantile}")







```

```python

#N_bulk
# Calculate Mean
mean = df['N_bulk'].mean()
# Calculate standard deviation
std = df['N_bulk'].std()
# Calculate quantile
quantile =df['N_bulk'].quantile()


print(f" Mean of N_bulk : {mean}")
print(f" standard deviation of N_bulk : {std}")
print(f" Quantile of N_bulk : {quantile}")
```
